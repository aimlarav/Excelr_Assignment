### Descriptive Analytics and Data Preprocessing on Sales & Discounts Dataset


## Descriptive Analytics for Numerical Columns

```python
# Load the dataset: Use Python and libraries like pandas to load the dataset:
import pandas as pd
data = pd.read_csv(r"D:\excelr_assignment\Basic stats - 1\sales_data_with_discounts.csv")
data
```

```python
# Identify numerical columns: Check for numerical columns:
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
numerical_cols
```

```python
# Calculate the mean, median, mode, and standard deviation for these columns.
stats = data[numerical_cols].describe().T
stats
```

```python
#mean
data[numerical_cols].mean()
```

```python
#median
data[numerical_cols].median()
```

```python
#mode
data[numerical_cols].mode()
```

```python
#standard deviation
data[numerical_cols].std()
```

## Data Visualization

```python
#Histograms:
#Plot histograms for each numerical column.
import matplotlib.pyplot as plt
data[numerical_cols].hist(figsize=(10,8))
plt.show()
```

```python
# Boxplot
data[numerical_cols].boxplot(figsize=(10,8))
plt.show()
```

```python
# Bar Charts for Categorical Columns:
categorical_cols = data.select_dtypes(include=['object']).columns
categorical_cols
```

```python
# Plot bar charts for each categorical variable:
for col in categorical_cols:
    data[col].value_counts().plot(kind='bar')
    plt.title(f'categorical distribution for {col}')
    plt.show()
```

## Standardization of Numerical Variables


```python
import seaborn as sns
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()
data_stanardized = data.copy()
data_stanardized[numerical_cols] = scaler.fit_transform(data[numerical_cols])
```

```python
print("Before Standardization:")
print(data[numerical_cols].describe())
```

```python
print("\nAfter Standardization:")
print(data_stanardized[numerical_cols].describe())
```

### Conversion of Categorical Data into Dummy Variables 

```python
data_encoded = pd.get_dummies(data_stanardized, columns=categorical_cols)

print("\nTransformed Dataset (Preview):")
print(data_encoded.head())
```
